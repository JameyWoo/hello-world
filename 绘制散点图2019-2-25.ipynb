```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn-whitegrid')
```

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o', color='red')
```

```python
rng = np.random.RandomState(42)
for maker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), maker, label="maker='{0}'".format(maker))
    plt.legend(numpoints=2) # 为了帮助我们展示每个数据对应的图像名称，更好的让读者认识到你的数据结构。
    plt.xlim(0, 1.8)
```

```python
plt.plot(x, y, '-ok', color='green')
```

```python
plt.plot(x, y, '-p', color='gray', markersize=12, linewidth=3, markerfacecolor='yellow', markeredgecolor='green', markeredgewidth=2)
plt.ylim(-1.2, 1.2) # 上下限
plt.xlim(12, -1) # 可逆序设置
```

```python
plt.plot(x, np.sin(x), '-o')
plt.axis([-1, 11, -1.2, 1.2]) # 一行代码设置区间
```

```python
plt.plot(x, np.sin(x), '-o')
plt.axis('tight') # 收紧
```

```python
plt.plot(x, np.sin(x), '-o')
plt.axis('equal') # 分辨率比例相等
```

```python
# 绘制指定大小的图
fig = plt.figure(figsize=(8, 5))
plt.plot(x, y, '-xk', color='blue')
```

```python
rng = np.random.RandomState(42)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)
# print(colors)
plt.scatter(x, y, c=colors, s=sizes, alpha=0.4, cmap='viridis') # alpha 调整透明度 cmap: 颜色图谱（colormap), 默认绘制为RGB(A)颜色空间。
plt.colorbar() # 显示颜色条
```

```python
# randn 与 rand
rn = np.random.RandomState(0)
print(rn.rand(10)) # [0: 1]中随机
print(rn.randn(10)) # 标准正态分布
```
